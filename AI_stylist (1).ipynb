```python
## One Hot Encoding converts into binary format
df = pd.get_dummies(df, columns=['gender', 'masterCategory', 'season', 'usage'])
df.head()
```

```python
## Lqbel Encoding for the ordinal relationship between the variables

from sklearn.preprocessing import LabelEncoder

# Label Encoding
label_encoders = {}
for column in ['subCategory', 'articleType', 'baseColour']:
    le = LabelEncoder()
    df[column] = le.fit_transform(df[column])
    label_encoders[column] = le

df
```

```python
## Target Encoding replaces the each category with the mean
import category_encoders as ce

np.random.seed(0)  # For reproducibility
df['target'] = np.random.randint(0, 2, size=len(df))

# Use target encoding for specified columns
encoder = ce.TargetEncoder(cols=['subCategory', 'articleType', 'baseColour'])
df_encoded = encoder.fit_transform(df, df['target'])

# Displaying the encoded DataFrame for comparison

df_encoded.head()
```

```python
# Frequency Encoding to calculate the frequaency of each category
for column in ['subCategory', 'articleType', 'baseColour']:
    freq = df[column].value_counts() / len(df)
    df[column + '_freq'] = df[column].map(freq)

df.head()
```

#### Feature Scaling

```python
## Normalization
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
numeric_columns = ['subCategory', 'articleType', 'baseColour']
scaled_features = scaler.fit_transform(df[numeric_columns])
scaled_df = pd.DataFrame(scaled_features, columns=[col + '_scaled' for col in numeric_columns])

# Concatenating scaled features back to the DataFrame
df = pd.concat([df, scaled_df], axis=1)

df
```

```python
# Fill missing values
df.fillna(df.median(), inplace=True)  # For numerical columns
df.fillna('Unknown', inplace=True) 
```

```python
# Dropping the original and non-numeric columns for PCA
from sklearn.decomposition import PCA
columns_to_drop = numeric_columns + ['productDisplayName', 'target']
pca_input = df.drop(columns=columns_to_drop)

# Applying PCA
pca = PCA(n_components=2)  # Adjust number of components as needed
reduced_features = pca.fit_transform(pca_input)
reduced_df = pd.DataFrame(reduced_features, columns=['PC1', 'PC2'])

# Adding PCA components back to the DataFrame
final_df = pd.concat([df, reduced_df], axis=1)

# Displaying the final DataFrame
final_df
```

```python
final_df.fillna(final_df.median(), inplace=True)  # For numerical columns
```
